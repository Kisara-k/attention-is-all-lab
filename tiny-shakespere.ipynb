```python
with open('input.txt', 'r', encoding='utf-8') as f:
    text = f.read()

len(text)
```

```python
chars = set(text)
chars = sorted(list(chars))

print(f'vocab {len(chars)}')
```

```python
s_to_i = { c: i for i, c in enumerate(chars) }
i_to_s = { i: c for i, c in enumerate(chars) }

encoder = lambda S: [s_to_i[c] for c in S]
decoder = lambda I: [i_to_s[i] for i in I]

print(encoder("What you egg"))
print(decoder(encoder("What you egg")))
```

```python
from tiktoken import get_encoding
enc = get_encoding('gpt2')
```

```python
print(f'vocab {enc.n_vocab}')

print(enc.encode("What you egg"))
print(enc.decode(enc.encode("What you egg")))
```

```python
# Encoder directly encodes each word, and does not apply stemming or lemmitization

words = " ".join([
    "running", "ran", "runs",
    "better", "good",
    "flying", "flew", "flies"
])
print(enc.encode(words))

words = " ".join([".", "happy.", "HAPPY"])
print(enc.encode(words))
```
